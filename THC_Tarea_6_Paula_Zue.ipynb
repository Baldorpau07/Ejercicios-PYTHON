```python
import numpy as np

#Funciones

def suma(matriz_1, matriz_2): #función de suma de matrices
  return np.add(matriz_1, matriz_2)

def producto(matriz_1, matriz_2): #función de multiplicación de matrices
  producto= np.dot(matriz_1, matriz_2)
  return producto

def traza(matriz_1): #función de traza de matrices
  traza= np.trace(matriz_1)
  return traza

def transpuesta(matriz_1): #función de transpuesta de matrices
  transpuesta= np.transpose(matriz_1)
  return transpuesta

def max_min(matriz_1): #función que encuentra el elemento mínimo y máximo de la matriz.
  maximo= float(np.max(matriz_1))
  minimo= float(np.min(matriz_1))
  return maximo, minimo

def suma_cuadrados(matriz_1): #función que suma los cuadrados de la matriz
  suma_cuadrados= np.sum(np.square(matriz_1))
  return suma_cuadrados

def matriz_alineada(matriz): #función que alinea la matriz
  print("[")
  for renglon in matriz:
    print(" ", renglon," ")
  print("]")
"""
matriz_alineada es una función estética que definí para que los renglones de la matriz queden alineados al imprimirlos.
En lugar de que se vea así:

La suma de las matrices es: [[10. 10. 10.]
 [ 9. 10.  9.]
 [ 5. 11. 16.]]

 Se vea así:

 [[10. 10. 10.]
 [ 9. 10.  9.]
 [ 5. 11. 16.]]
 """

def datos_renglon(texto): #función que filtra los datos que ingresa el usuario y anexa las columnas a la matriz correspondiente.
  while True:
    try:
      datos=list(map(float, input(texto).split()))
      if len(datos)==3:
        return datos
      else:
        print("Por favor ingrese exactamente 3 números reales (separados por espacios) en cada renglón.")
    except ValueError:
      print("Debe ingresar números reales.")

#Intrucciones para el usuario
print("Bienvenido a la calculadora de matrices 3x3")
print("Por favor ingrese exactamente 3 números reales (separados por espacios) en cada renglón")
print("Por ejemplo: 1 0 -4")
print("Deberá ingresar los datos de 2 matrices")
print("------------------------------------------------------------------------------------------------------------------------")
print("Datos de las matrices\n")

#Matriz 1
diccionario_matriz1={} #diccionario de matriz 1
diccionario_matriz1["x"]=datos_renglon("Ingrese el primer renglón de la matriz 1: ")
diccionario_matriz1["y"]=datos_renglon("Ingrese el segundo renglón de la matriz 1: ")
diccionario_matriz1["z"]=datos_renglon("Ingrese el tercer renglón de la matriz 1: ")
matriz_1=np.array([diccionario_matriz1["x"], diccionario_matriz1["y"], diccionario_matriz1["z"]]) #array de la matriz 1
print("\nLa matriz 1 es:")
matriz_alineada(matriz_1)

#Matriz 2
diccionario_matriz2={} #diccionario de matriz 2
diccionario_matriz2["a"]=datos_renglon("\nIngrese el primer renglón de la matriz 2: ")
diccionario_matriz2["b"]=datos_renglon("Ingrese el segundo renglón de la matriz 2: ")
diccionario_matriz2["c"]=datos_renglon("Ingrese el tercer renglón de la matriz 2: ")
matriz_2=np.array([diccionario_matriz2["a"], diccionario_matriz2["b"], diccionario_matriz2["c"]]) #array de la matriz 2
print("\nLa matriz 2 es:")
matriz_alineada(matriz_2)

print("\n------------------------------------------------------------------------------------------------------------------------")
print("Operaciones con matrices\n")

print("La suma de las matrices es:")
matriz_alineada(suma(matriz_1, matriz_2))

print("\nEl producto de las matrices es:")
matriz_alineada(producto(matriz_1, matriz_2))

print("\nLa traza de la matriz 1 es:", traza(matriz_1))
print("La traza de la matriz 2 es:", traza(matriz_2))

print("\nLa transpuesta de la matriz 1 es:")
matriz_alineada(transpuesta(matriz_1))
print("La transpuesta de la matriz 2 es:")
matriz_alineada(transpuesta(matriz_2))

print("\nEl valor máximo y mínimo de la matriz 1 es:", max_min(matriz_1))
print("El valor máximo y mínimo de la matriz 2 es:", max_min(matriz_2))

print("\nLa suma de los cuadrados de la matriz 1 es:", suma_cuadrados(matriz_1))
print("La suma de los cuadrados de la matriz 2 es:", suma_cuadrados(matriz_2))
```

```python
#Punto 7 de la primera parte de matrices

matriz_A=np.array([[11,7,-1],[-3,5,2],[4,2,-2]])
matriz_B=np.array([[1,1,1],[10,3,-5],[-6,-10,-3]])
print("Matriz A:\n")
print(matriz_A,"\n")
print("Matriz B:\n")
print(matriz_B)

print("\nInstrucciones:\n")
print("Calcule:")
print("2B+3A")
print("(B+2A)^2")
print("A^2 +3B")
print("\n------------------------------------------------------------------------------------------------------------------------\n")

print("Operaciones:\n")
def suma2(matriz_A, matriz_B): #función de suma de matrices
  suma= np.add(matriz_A, matriz_B)
  return suma

def matriz_cuadrada(M): #función que eleva al cuadrada una matriz
  matriz_cuadrada= np.dot(M,M)
  return matriz_cuadrada

def multiplicacion_escalar(matriz_A,escalar): #función que multiplica las entradas de una matriz por un escalar
  multiplicacion_escalar= np.multiply(matriz_A, escalar)
  return multiplicacion_escalar

#Resultados y llamadas a las funciones correspondientes
print("2B+3A:\n")
print(suma2(multiplicacion_escalar(matriz_B,2),multiplicacion_escalar(matriz_A,3)))

print("\n(B+2A)^2:\n")
print(matriz_cuadrada(suma2(matriz_B,multiplicacion_escalar(matriz_A,2))))

print("\nA^2 +3B:\n")
print(suma2(matriz_cuadrada(matriz_A),multiplicacion_escalar(matriz_B,3)))
```

```python

import numpy as np
def matriz_cuadrada(): #función que le pide al usuario el número de filas y columnas.
  while True:
    try:
      n=int(input("Ingrese el número de filas de la matriz: "))
      m=int(input("Ingrese el número de columnas de la matriz: "))
      if n<3 or m<3 or n>5 or m>5:
        print("Debe ingresar un número entre 3 y 5")
        continue
      if n!=m: #Error, por si el usuario ingresa un número de filas distinto al de columnas.
        print("La matriz no es cuadrada. Vuelva a ingresar los datos.")
        continue
      matriz_T = np.random.randint(0, 1000, size=(n, m)) #Llena la matriz cuadrada con entradas random.
      print("\nLa matriz T es:\n")
      print(matriz_T)
      return matriz_T
    except ValueError:
      print("Debe ingresar números enteros (positivos).")

def traza2(matriz): #función de traza
  traza= np.trace(matriz)
  return traza

def max_min2(matriz): #función de valores máximos y mínimos
  maximo= float(np.max(matriz))
  minimo= float(np.min(matriz))
  return maximo, minimo

#Instrucciones
print("\n------------------------------------------------------------------------------------------------------------------------")
print("Instrucciones:\n")
print("Ingrese el número de renglones y columnas de la matriz, de tal manera que sea cuadrada (para toda i,j con i=j 3<=i,j<=5)")
print("------------------------------------------------------------------------------------------------------------------------")
matriz_T=matriz_cuadrada() #llamada a la función

filas_pares=0
columnas_impares=0

#Suma de elementos pares recorriendo filas
for fila in matriz_T:
  for elemento in fila:
    if elemento%2==0: #módulo de pares
      filas_pares+=elemento

#Suma de elementos impares recorriendo columnas
for columna in matriz_T.T: #Transpuse la matriz para poder tratar las columnas como filas y hacer la suma más sencilla.
  for elemento in columna:
    if elemento%2!=0: #módulo de impares
      columnas_impares+=elemento

#-----------------------------------------------------------------------------------------------------

#Aquí imprimí los resultados
print("\n------------------------------------------------------------------------------------------------------------------------")
print("Operaciones:\n")

print("\n-Primera fila inversa:", matriz_T[0,::-1]) #Toma la primera fila, todos los valores y los inverte en orden (con el -1).
print("\n-Primera columna inversa:", matriz_T[::-1,0]) #Toma las filas, las invierte en orden, y toma la primera columna.
print("\n-Primer elemento y último elemento:", matriz_T[0,0], matriz_T[-1,-1])

print("\n-Valor máximo y mínimo", max_min2(matriz_T))
print("\n-Promedio de la diagonal",round(np.mean(np.diagonal(matriz_T)),3))
print("\n-Traza", traza2(matriz_T))

print("\n-Suma de los elementos pares recorriendo las filas=", filas_pares)
print("-Suma de los elementos impares recorriendo las columnas =", columnas_impares)

promedio_filas=np.round(np.mean(matriz_T, axis=1),3) #Promedio x filas, redondeado a 3 decimales.
print("\n-Promedio por fila:")
for z,valor in enumerate(promedio_filas,start=1): #Me enumera los promedio x filas para que se vea más ordenado.
  print(f"Fila {z}: {valor}")

promedio_columnas=np.round(np.mean(matriz_T, axis=0),3) #Promedio x columnas, redondeado a 3 decimales.
print("\n-Promedio por columna:")
for z,valor in enumerate(promedio_columnas,start=1): #Me enumera los promedio x columnas para que se vea más ordenado.
  print(f"Columna {z}: {valor}")

print("\n-Antidiagonal:", np.diagonal(np.fliplr(matriz_T)))
print("\n-Promedio de la antidiagonal:", round(np.mean(np.diagonal(np.fliplr(matriz_T))),3))
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-light.mplstyle')
#Nota: el estilo lo saqué de la página de matplotlib.


#Gráfica de ecuación diferencial

x=np.linspace(-10,10,400)
font1 = {'family':'serif','color':'black','fontsize':15} #Diccionario de tipografía

plt.figure(figsize=(6, 6))

c_valores=[-3,0,3] #Valores de C utilizados para graficar

c1=c_valores[0]
y1=(x-1)+(c1*np.exp(-x))
plt.plot(x,y1,label=f"Valores de C<0",color='rebeccapurple')

c2=c_valores[1]
y2=(x-1)+(c2*np.exp(-x))
plt.plot(x,y2,label=f"Valores de C=0",color='darkmagenta')

c3=c_valores[2]
y3=(x-1)+(c3*np.exp(-x))
plt.plot(x,y3,label=f"Valores de C>0",color='violet')
plt.legend(frameon=True,edgecolor='black') #Recuadro de leyendas

plt.title("Familia de curvas", fontdict=font1,y=1.07) #y modifica la altura del título y subtítulo.
plt.suptitle("y=(x-1)+C e^(-x)", fontdict=font1,y=1.045)

plt.xlabel("Eje de las x")
plt.ylabel("Eje de las y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-dark.mplstyle')
#Nota: el estilo lo saqué de la página de matplotlib.

font1 = {'family':'serif','color':'lavender','fontsize':15} #Diccionario de tipografía
plt.figure(figsize=(6, 6))

z=np.linspace(-10,10,1001) #Rango de -10 a 10 con más de 1000 muestras
p=(3*z+7)/((z**2)-4) #Función a graficar

plt.plot(z,p,label=f"Función",color='lavender')

#Asíntotas verticales
plt.axvline(x=2,color='blueviolet',linestyle='--',label='Asíntota x=2')
plt.axvline(x=-2,color='blueviolet',linestyle='--',label='Asíntota x=-2')

#Asíntota horizontal
plt.axhline(y=0,color='r',linestyle='--',label='Asíntota y=0')

plt.xlabel("Eje de las x")
plt.ylabel("Eje de las y")
plt.title("Gráfica de la función", fontdict=font1,y=1.04)

plt.xlim(-10,10) #límites de x que van de -10 a 10
plt.ylim(-10,10) #límites de y que van de -10 a 10

plt.legend(frameon=True,edgecolor='black') #Cuadrito de leyendas
plt.show()
```
