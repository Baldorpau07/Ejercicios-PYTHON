```python
#Ejercicio de programa de sumas Leetcode con input
print("Programa de sumas")
print("El siguiente programa suma a_i+a_(i-1) con a_1 com valor inicial\n")
print("Ingrese 4 números enteros\n")

Solution=[]
while True:
  try:
    while len(Solution)<4:
      Solution.append(int(input()))
    break
  except ValueError:
    print("Ingrese un número entero")
for i in range(len(Solution)):
  if i==0:
    Solution[i] = Solution[0]
  else:
    Solution[i] = Solution[i-1] + Solution[i]
print("\nSuma de números ingresados:",Solution)
```
