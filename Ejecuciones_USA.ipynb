```python
victimas_por_raza = {}

for i, fila in df.iterrows(): #Esto lo investigu, iterrows se usa para recorrer la base de datos fila por fila.
    raza = fila["Victim Race"]
    num  = fila["Victim Count"]

    if raza not in victimas_por_raza:
        victimas_por_raza[raza] = 0

    victimas_por_raza[raza] += num

victimas_por_raza

#Con estos datos que conseguimos el total de vctimas, visto que 174 no tienen raza asignada, de los cuales 168 son del atentado de Oklahoma.
```

```python
#Este cdigo es para contar las vctimas mixtas para despus sumrselo al total de vcitimas por raza.
mixtos=df[df['Victim Race'].notna() & df['Victim Race'].str.contains(',',na=False)]
victimas_mixtas = {
    "White": 0,
    "Black": 0,
    "Latino": 0,
    "Asian": 0,
    "Native American": 0
}

for _, fila in mixtos.iterrows():
    desc = fila["Victim Race"]   # Ej: "4 White, 1 Black"

    partes = desc.split(",")     # ["4 White", "1 Black"]

    for parte in partes:
        parte = parte.strip()
        trozos = parte.split()   # ["4","White"] o ["White"]

        # Caso 1: tiene nmero explcito
        if trozos[0].isdigit():
            n = int(trozos[0])
            raza = " ".join(trozos[1:])
        else:
            # Caso raro: no trae nmero  asumir 1
            n = 1
            raza = parte

        # Sumar al diccionario
        if raza in victimas_mixtas:
            victimas_mixtas[raza] += n
victimas_mixtas
```

```python
#Porcentajes por raza con respecto al total
w=round((vrace_final['White']*100/total),2)
b=round((vrace_final['Black']*100/total),2)
a=round((vrace_final['Asian']*100/total),2)
l=round((vrace_final['Latino']*100/total),2)
d=round((vrace_final['nan']*100/total),2)
n=round((vrace_final['Native American']*100/total),2)
print(w,b,a,l,d,n)
```
