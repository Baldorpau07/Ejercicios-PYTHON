```python
#slicing
#se puede definir [inicio:fin]
#o se puede definir con el paso [inicio:fin:paso]
```

```python
arr2=np.array((1,2,3,4,5))
print(arr2[1:4])
#aquí se usa slicing en el qie imprime los que tú quieras poner [inicio:fin] pero curiosamente el fin es como una cota superior que no imprime, es el tope
print(arr2[1:4:2])
#este es el paso que sirve para saltarse alguna posición al imprimir
print(arr2[1:])
#este slicing es para que imprima desde la primera posición hasta el final (los sucesores que están en el arr2)
print(arr2[:4])
#imprime todos los antecesores hasta la posición 4
```

```python
arr3=np.array([[1,2,3],[4,5,6]])
print(arr3[0,:2])
#aquí le está diciendo a numpy, ve al primer vector e imprime todos sus elementos anteriores a la posición 2
arr4=np.array([[[1,2,3],[4,5,6]],[[1,2,4],[4,5,6]]])
print(arr4[0:2,0:2,0:2])
#ahí imprimió los primeros dos elementos de cada fila
```

```python
arr8=np.array((1,2,3,4,5,6,7,8))
x=np.where(arr8==3)
#me da la el elemento en la posición 3 en el arr8
y=np.where(arr8%2==0)
#me da las posiciones de los números que son divisibles entre 2
print(x)
print(y)
z=np.searchsorted(arr8, 7)
print(z)
#searchsorted es más específico, en este caso busca la posiciíon del 7
```

```python
#Filtrado
arr8=np.array((1,2,3,4,5,6,7,8))
filter_arr=arr8 >4
#aquellos que cumplen con la condición de ser mayores a 4
newarr=arr8[filter_arr]
print(filter_arr)
#imprime true quienes cumplen con la condición y false los que no (un buleano)
print(newarr)
#imprime los que cumplen con la condición
```
